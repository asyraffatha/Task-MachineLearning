```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline
sns.set(color_codes=True)
```

```python
# Membaca file CSV yang diunggah ke dalam DataFrame
# Gunakan nama file yang sebenarnya dari hasil upload
filename = list(uploaded.keys())[0]  # Mendapatkan nama file yang diunggah
df = pd.read_csv(filename)
```

```python
# Menampilkan 5 baris pertama
df.head(5)
```

```python
#Menampilkan 5 baris terakhir
df.tail(5)
```

```python
#Menampilkan tipe data
df.dtypes
```

```python
#Mengahapus kolom 'Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', dan 'Vehicle Size'.
#Menampilkan kembali 5 baris pertama setelah kolom-kolom tersebut dihapus
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

```python
#Mengganti nama kolom menjadi lebih singkat
df = df.rename(columns={"Engine HP": "HP",
                        "Engine Cylinders": "Cylinders",
                        "Transmission Type": "Transmission",
                        "Driven_Wheels": "Drive Mode",
                        "highway MPG": "MPG-H",
                        "city mpg": "MPG-C",
                        "MSRP": "Price"})
df.head(5)
```

```python
#Mengetahui ukuran dataframe
df.shape
```

```python
#Mengetahui jumlah baris duplikat
duplicate_rows_df = df[df.duplicated()]
print("Number of duplicate rows: ", duplicate_rows_df.shape[0])
```

```python
#Menghitung jumlah nilai yang tidak kosong
df.count()
```

```python
#Menghapus baris duplikat
df = df.drop_duplicates()
df.head(5)
```

```python
#Menghitung kembali jumlah nilai yang tidak kosong
df.count()
```

```python
#Menghitung dan menampilkan jumlah nilai yang hilang
print(df.isnull().sum())

#Menghapus semua baris yg memili nilai kosong
df = df.dropna()
df.count()

```

```python
sns.boxplot(x=df['Price'])
```

```python
sns.boxplot(x=df['HP'])
```

```python
sns.boxplot(x=df['Cylinders'])
```

```python
#Asumsi 'df' adalah dataframe

#Mengubah kolom yang relevan menjadi numerik, mengatasi kesalahan
for column in ['Price', 'HP', 'Cylinders']: #Menyertakan semua kolom yg relevan
    try:
        df[column] = pd.to_numeric(df[column], errors='coerce')
        # 'coerce' akan mengganti nilai non-numerik dengan NaN
    except (TypeError, ValueError):
        print(f"Could not convert column '{column}' to numeric.")

#Menghitung kuantil
Q1 = df.quantile(0.25, numeric_only=True) # Tentukan numeric_only
Q3 = df.quantile(0.75, numeric_only=True)
IQR = Q3 - Q1
print(IQR)
```

```python
# Menghapus outliers
# Using .loc to select the relevant columns when comparing
filtered_df = df[~((df.loc[:, ['Price', 'HP', 'Cylinders']] < (Q1[['Price', 'HP', 'Cylinders']] - 1.5 * IQR[['Price', 'HP', 'Cylinders']])) | (df.loc[:, ['Price', 'HP', 'Cylinders']] > (Q3[['Price', 'HP', 'Cylinders']] + 1.5 * IQR[['Price', 'HP', 'Cylinders']]))).any(axis=1)]
# Menampilkan bentuk DataFrame setelah pembersihan
print(filtered_df.shape)
```

```python
#Menghitung dan memvisualisasikan jumlah mobil berdasarkan merek (make)
# Visualisasi jumlah mobil berdasarkan merek
df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make')
plt.show()
```

```python
#Membuat dan menampilkan heatmap dari matriks korelasi
plt.figure(figsize=(10,5))
# Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=['number'])
c = numeric_df.corr()
sns.heatmap(c, cmap="BrBG", annot=True)
plt.title("Correlation Heatmap")
plt.show()
c  # Menampilkan matriks korelasi
```

```python
# Membuat scatter plot
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['HP'], df['Price'])
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.title("Scatter Plot of HP vs Price")
plt.show()
```
