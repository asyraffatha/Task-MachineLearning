Beri sample kode python untuk dapat menampilkan persentase dari data bertipe kategorikal dalam bentuk pie chart, sehingga saya bisa copy paste ke google colab.

````python
# Import libraries yang dibutuhkan
import pandas as pd
import matplotlib.pyplot as plt

# Baca dataset
file_path = '/content/drive/MyDrive/StudentsPerformance.csv'
df = pd.read_csv(file_path)

# Fungsi untuk membuat pie chart dari kolom bertipe kategorikal
def plot_pie_chart(column_name):
    # Hitung persentase setiap kategori
    data = df[column_name].value_counts()

    # Buat pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Distribution of {column_name}')
    plt.show()

# Contoh penggunaan untuk kolom 'gender'
plot_pie_chart('gender')

# Kamu juga bisa mencoba kolom lain, misalnya:
plot_pie_chart('race/ethnicity')
plot_pie_chart('parental level of education')
plot_pie_chart('lunch')
plot_pie_chart('test preparation course')
````

Beri kode python untuk melihat jumlah siswa yang fail dan pass untuk masing-masing pelajaran (math, reading, writing score). Dengan minimal nilai yang dianggap pass adalah 75!

````python
import pandas as pd

# Misalkan df adalah DataFrame yang sudah ada
# Contoh struktur DataFrame
# df = pd.read_csv('StudentsPerformance.csv')

def count_pass_fail(df):
    # Menghitung jumlah siswa yang lulus dan gagal untuk setiap pelajaran
    pass_fail_counts = {
        'Math': {
            'Pass': (df['math score'] >= 75).sum(),
            'Fail': (df['math score'] < 75).sum()
        },
        'Reading': {
            'Pass': (df['reading score'] >= 75).sum(),
            'Fail': (df['reading score'] < 75).sum()
        },
        'Writing': {
            'Pass': (df['writing score'] >= 75).sum(),
            'Fail': (df['writing score'] < 75).sum()
        }
    }

    return pass_fail_counts

# Menampilkan jumlah siswa yang pass dan fail
pass_fail_results = count_pass_fail(df)

# Menampilkan hasil
for subject, counts in pass_fail_results.items():
    print(f"{subject}: Pass = {counts['Pass']}, Fail = {counts['Fail']}")
````

Prompt: Beri kode untuk melihat nilai rata-rata seluruh mata pelajaran (math, reading writing) dari setiap siswa dan lihat korelasinya dengan masing-masing kategori

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Misalkan df adalah DataFrame yang sudah ada
# df = pd.read_csv('StudentsPerformance.csv')

# Menambahkan kolom rata-rata nilai
df['average score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

# Fungsi untuk menampilkan rata-rata nilai berdasarkan kategori
def plot_average_score_by_category(df, category):
    plt.figure(figsize=(10, 6))

    # Boxplot untuk nilai rata-rata
    sns.boxplot(data=df, x=category, y='average score', palette='pastel')
    plt.title(f'Average Score by {category.capitalize()}', fontsize=16)
    plt.ylabel('Average Score')
    plt.xticks(rotation=45)

    plt.show()

# Menampilkan visualisasi untuk setiap kategori
categories = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

for category in categories:
    plot_average_score_by_category(df, category)

# Mengonversi kolom kategorikal menjadi variabel numerik menggunakan one-hot encoding
df_encoded = pd.get_dummies(df, columns=categories, drop_first=True)

# Menampilkan korelasi antara nilai rata-rata dan kategori
correlation_matrix = df_encoded.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix', fontsize=16)
plt.show()
````

Tambahan visualisasi
Pairplot: Untuk melihat hubungan antara nilai math, reading, dan writing secara bersamaan berdasarkan kategori.
Heatmap: Untuk menunjukkan korelasi antara nilai mata pelajaran.
Countplot: Untuk melihat distribusi jumlah siswa berdasarkan kategori seperti gender dan lunch.
Violin Plot: Untuk membandingkan distribusi nilai berdasarkan kategori dengan cara yang lebih informatif.

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Misalkan df adalah DataFrame yang sudah ada
# df = pd.read_csv('StudentsPerformance.csv')

# Mengatur gaya visualisasi
sns.set(style="whitegrid")

# 1. Pairplot untuk melihat hubungan antar mata pelajaran
def plot_pairplot(df):
    plt.figure(figsize=(10, 8))
    sns.pairplot(df, vars=['math score', 'reading score', 'writing score'], hue='gender', palette='pastel')
    plt.suptitle('Pairplot of Scores by Gender', y=1.02)
    plt.show()

# 2. Heatmap untuk menunjukkan korelasi antara nilai mata pelajaran
def plot_heatmap(df):
    plt.figure(figsize=(8, 6))
    correlation = df[['math score', 'reading score', 'writing score']].corr()
    sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar_kws={"shrink": .8})
    plt.title('Correlation Heatmap of Scores', fontsize=16)
    plt.show()

# 3. Countplot untuk melihat distribusi gender
def plot_countplot(df, category):
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=category, palette='pastel')
    plt.title(f'Distribution of {category.capitalize()}', fontsize=16)
    plt.ylabel('Number of Students')
    plt.xticks(rotation=45)
    plt.show()

# 4. Violin plot untuk distribusi nilai berdasarkan kategori
def plot_violin(df, subject, category):
    plt.figure(figsize=(12, 6))
    sns.violinplot(data=df, x=category, y=subject, palette='pastel', inner='quartile')
    plt.title(f'Distribution of {subject.capitalize()} Scores by {category.capitalize()}', fontsize=16)
    plt.ylabel(f'{subject.capitalize()} Score')
    plt.xticks(rotation=45)
    plt.show()

# Memanggil fungsi visualisasi
plot_pairplot(df)
plot_heatmap(df)

# Memanggil countplot untuk kategori gender dan lunch
plot_countplot(df, 'gender')
plot_countplot(df, 'lunch')

# Memanggil violin plot untuk setiap mata pelajaran berdasarkan gender
for subject in ['math score', 'reading score', 'writing score']:
    plot_violin(df, subject, 'gender')
````



1.   Pairplot

*   Hubungan Antara Mata Pelajaran: Pairplot menunjukkan hubungan antara nilai math, reading, dan writing. digunakan untuk mengamati bahwa siswa yang berprestasi tinggi di satu mata pelajaran cenderung juga berprestasi baik di mata pelajaran lainnya. Misalnya, siswa dengan nilai tinggi di math juga sering memiliki nilai tinggi di reading dan writing.
*   Dampak Gender: Dengan menggunakan pewarnaan berdasarkan gender, kita mungkin melihat perbedaan dalam distribusi nilai antara siswa laki-laki dan perempuan, yang bisa memberikan wawasan tentang performa akademik berdasarkan gender.

2.   Heatmap

*   Korelasi Mata Pelajaran: Heatmap menampilkan koefisien korelasi antara nilai mata pelajaran. Nilai yang lebih mendekati 1 atau -1 menunjukkan korelasi yang lebih kuat. Misalnya, jika math score dan reading score memiliki korelasi tinggi, itu menunjukkan bahwa siswa yang baik dalam matematika juga cenderung baik dalam membaca.
*   Pengambilan Keputusan: Pemahaman ini dapat membantu pendidik dalam merancang kurikulum yang saling terkait antara mata pelajaran.

3. Countplot

*   Distribusi Gender: Countplot untuk gender menunjukkan jumlah siswa laki-laki dan perempuan. Jika jumlah siswa perempuan lebih banyak dibandingkan laki-laki, ini dapat menunjukkan ketertarikan atau akses yang lebih baik bagi perempuan dalam pendidikan di area tersebut.
*  Distribusi Jenis Makanan: Countplot untuk lunch menunjukkan jumlah siswa yang mengikuti program makanan gratis/reduksi dibandingkan dengan yang membayar penuh. Jika ada lebih banyak siswa yang mendapatkan makanan gratis, itu dapat menunjukkan bahwa banyak siswa berasal dari latar belakang ekonomi yang lebih rendah

4. Violin Plot

*   Distribusi Nilai Berdasarkan Kategori: Violin plot menunjukkan distribusi nilai math, reading, dan writing berdasarkan kategori seperti gender. Dengan melihat bentuk violin, kita dapat memahami di mana nilai terkonsentrasi dan apakah ada perbedaan yang signifikan antara laki-laki dan perempuan.
*   Median dan Kuartil: Violin plot juga menampilkan median dan kuartil, yang dapat memberikan pemahaman yang lebih baik tentang nilai pusat dan rentang nilai yang diperoleh siswa. Misalnya, jika siswa perempuan cenderung memiliki median yang lebih tinggi di writing, ini menunjukkan bahwa mereka lebih unggul dalam keterampilan menulis dibandingkan laki-laki.
