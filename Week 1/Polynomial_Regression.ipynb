```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
# Upload file
from google.colab import files
uploaded = files.upload()
```

```python
# Load the dataset
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
# Check the dataset
print(dataset.head())
```

```python
# Select the features (Level) and target variable (Salary)
X = dataset.iloc[:, 1:2].values  # Assuming 'Level' is the second column
y = dataset.iloc[:, 2].values    # Assuming 'Salary' is the third column
```

```python
# Train a simple linear regression model for comparison
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# Train a Polynomial Regression model (degree 4 as example)
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
lin_reg_poly = LinearRegression()
lin_reg_poly.fit(X_poly, y)
```

```python
# Visualizing the Linear Regression results
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg.predict(X), color='blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualizing the Polynomial Regression results
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg_poly.predict(poly_reg.fit_transform(X)), color='blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# Predicting a new result with Linear Regression
linear_prediction = lin_reg.predict([[6.5]])
```

```python
# Predicting a new result with Polynomial Regression
poly_prediction = lin_reg_poly.predict(poly_reg.fit_transform([[6.5]]))
```

```python
# Print predictions
print(f"Linear Regression Prediction for 6.5: {linear_prediction}")
print(f"Polynomial Regression Prediction for 6.5: {poly_prediction}")
```

```python
# Calculate and print mean squared error for both models
mse_linear = mean_squared_error(y, lin_reg.predict(X))
mse_poly = mean_squared_error(y, lin_reg_poly.predict(X_poly))

print(f"Mean Squared Error (Linear): {mse_linear}")
print(f"Mean Squared Error (Polynomial): {mse_poly}")
```
